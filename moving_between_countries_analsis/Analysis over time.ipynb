```python
import pandas as pd
import matplotlib.pyplot as plt

# Input file path
input_file = '../data/moving_between_countries/transitions_to_from_israel.csv'

# Define the minimum year for plotting
min_year = 2000  # Set your desired minimum year here

# Read the input CSV file
df = pd.read_csv(input_file)

# Filter the data based on the minimum year
df_filtered = df[df['year'] >= min_year]

# Set the figure size
plt.figure(figsize=(12, 7))

# Plotting side-by-side bar chart for left and arrived transitions
bar_width = 0.35  # Width of the bars
index = df_filtered['year']  # Years to be used as x-axis positions

# Plotting the bars for 'left' transitions (negative for visual clarity)
plt.bar(index - bar_width / 2, df_filtered['left_permanent'], bar_width, label='Left Permanent', color='r')
plt.bar(index + bar_width / 2, df_filtered['arrived_permanent'], bar_width, label='Arrived Permanent', color='b')
plt.bar(index - bar_width / 2, df_filtered['left_temporary'], bar_width, bottom=df_filtered['left_permanent'], label='Left Temporary', color='g')
plt.bar(index + bar_width / 2, df_filtered['arrived_temporary'], bar_width, bottom=df_filtered['arrived_permanent'], label='Arrived Temporary', color='y')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Number of Transitions')
plt.title(f'Comparing Left and Arrived Transitions per Year (After {min_year})')

# Add legend, grid, and adjust layout
plt.legend(title='Transition Type')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Input file path
input_file = '../data/moving_between_countries/transitions_to_from_israel.csv'

# Define the minimum year for plotting
min_year = 1978  # Set your desired minimum year here

# Read the input CSV file
df = pd.read_csv(input_file)

# Filter the data based on the minimum year
df_filtered = df[df['year'] >= min_year]

# Calculate the difference between arrivals and departures for both permanent and temporary
df_filtered['permanent_gap'] = df_filtered['arrived_permanent'] - df_filtered['left_permanent']
df_filtered['temporary_gap'] = df_filtered['arrived_temporary'] - df_filtered['left_temporary']

# Set up the plot
plt.figure(figsize=(12, 7))

# Plot the gaps for permanent and temporary arrivals vs departures
plt.bar(df_filtered['year'] - 0.2, df_filtered['permanent_gap'], width=0.4, label='Permanent Arrival - Departure Gap', color='b')
plt.bar(df_filtered['year'] + 0.2, df_filtered['temporary_gap'], width=0.4, label='Temporary Arrival - Departure Gap', color='y')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Gap (Arrival - Departure)')
plt.title(f'Gap Between Arrivals and Departures (After {min_year})')

# Add legend, grid, and adjust layout
plt.legend(title='Gap Type')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Input file path
input_file = '../data/moving_between_countries/transitions_to_from_israel.csv'

# Define the minimum and maximum year for plotting
min_year = 1948
max_year = 2021

# Larger font size parameters
fontsize = 20       # For labels, title, legend
tick_fontsize = 18  # For axis tick labels
legend_fontsize = 18
title_fontsize = 24

# Read the input CSV file
df = pd.read_csv(input_file)

# Filter the data based on the minimum and maximum year
df_filtered = df[(df['year'] >= min_year) & (df['year'] <= max_year)]

# Calculate the gap between arrivals and departures for both permanent and temporary
df_filtered['permanent_gap'] = df_filtered['arrived_permanent'] - df_filtered['left_permanent']
df_filtered['temporary_gap'] = df_filtered['arrived_temporary'] - df_filtered['left_temporary']

# Calculate the total arrivals and total departures for each year
df_filtered['total_arrivals'] = df_filtered['arrived_permanent'] + df_filtered['arrived_temporary']
df_filtered['total_departures'] = df_filtered['left_permanent'] + df_filtered['left_temporary']

# Calculate the difference between total arrivals and total departures
df_filtered['arrival_departure_gap'] = df_filtered['total_arrivals'] - df_filtered['total_departures']

# Set up the plot
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot the gap between permanent and temporary arrivals and departures (scatter plot)
scatter1 = ax1.scatter(df_filtered['year'], df_filtered['permanent_gap'], label='Permanent Gap (Arrival - Departure)', color='b', marker='o', s=80)
scatter2 = ax1.scatter(df_filtered['year'], df_filtered['temporary_gap'], label='Temporary Gap (Arrival - Departure)', color='g', marker='x', s=80)

# Create a second y-axis for the bar chart of total arrivals vs total departures
ax2 = ax1.twinx()
bar = ax2.bar(df_filtered['year'], df_filtered['arrival_departure_gap'], label='Total Gap (Arrivals - Departures)', color='r', alpha=0.5)

# Adjust the range of the y-axis with padding
min_val = min(df_filtered['permanent_gap'].min(), df_filtered['temporary_gap'].min(), df_filtered['arrival_departure_gap'].min())
max_val = max(df_filtered['permanent_gap'].max(), df_filtered['temporary_gap'].max(), df_filtered['arrival_departure_gap'].max())
padding = (max_val - min_val) * 0.05  # 5% padding

ax1.set_ylim(min_val - padding, max_val + padding)
ax2.set_ylim(min_val - padding, max_val + padding)

# Labels and title with increased font size
ax1.set_xlabel('Year', fontsize=fontsize)
ax1.set_ylabel('Gap (Arrival - Departure)', color='black', fontsize=fontsize)
ax2.set_ylabel('Total Gap (Arrivals - Departures)', color='r', fontsize=fontsize)
plt.title(f'Gaps Between Arrivals and Departures in Israel (From {min_year} to {max_year})', fontsize=title_fontsize)

# Set tick label sizes
ax1.tick_params(axis='x', labelsize=tick_fontsize)
ax1.tick_params(axis='y', labelsize=tick_fontsize)
ax2.tick_params(axis='y', labelsize=tick_fontsize)

# Combine all handles and labels for a single legend
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
all_handles = handles1 + handles2
all_labels = labels1 + labels2

# Place one combined legend below the plot
plt.legend(all_handles, all_labels, title='Gap Type',
           bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=2,
           fontsize=legend_fontsize, title_fontsize=legend_fontsize)

# Add grid and adjust layout
ax1.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
