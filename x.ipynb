```python
import pandas as pd
from collections import Counter

# path to your large CSV file
csv_path = "data/merged_openalex_works_data3500.csv"

# Counter to accumulate counts per year
year_counts = Counter()

# process the file in chunks
for chunk in pd.read_csv(
    csv_path,
    usecols=["publication_year"],  # load only the year column
    chunksize=100_000,             # adjust chunk size to your memory limits
    low_memory=True
):
    # count occurrences in this chunk and update the global counter
    year_counts.update(chunk["publication_year"].value_counts().to_dict())

# אם רוצים תוצאה מסודרת לפי שנה:
counts_df = (
    pd.DataFrame.from_dict(year_counts, orient="index", columns=["works_count"])
      .rename_axis("publication_year")
      .sort_index()
)

print(counts_df.to_string())
```

```python
import pandas as pd
from collections import Counter

# Path to your large CSV file
FILENAME = "data/merged_openalex_works_data3500.csv"

# Column name to aggregate by
COUNTRY_COL = "country"

# Number of rows per chunk (אפשר לכוונן לפי זיכרון הזמין)
CHUNKSIZE = 2_000_000

# Counter לאגירת הספירות
country_counts = Counter()

# קוראים את הקובץ ב־chunks כדי לחסוך בזיכרון
for chunk in pd.read_csv(
    FILENAME,
    usecols=[COUNTRY_COL],     # טוענים רק את עמודת המדינה
    chunksize=CHUNKSIZE,
    low_memory=True
):
    # update the overall counts with this chunk's value_counts
    country_counts.update(chunk[COUNTRY_COL])

# ממירים ל־DataFrame כדי להציג בצורה מסודרת
counts_df = pd.DataFrame.from_dict(
    country_counts, 
    orient="index", 
    columns=["work_count"]
)

# sort descending by count
counts_df.sort_values("work_count", ascending=False, inplace=True)

# מדפיסים את התוצאות
print(counts_df.to_string())
```

```python
import pandas as pd

# Path to your large CSV file
FILENAME = "data/merged_openalex_works_data3500.csv"

# Output CSV for works with authors from multiple countries
OUTPUT_FILE = "multi_country_works.csv"

# Columns we need
COLS = ["work_id", "publication_year", "country"]

# Adjust chunksize to your available RAM (e.g. ~1e6 rows ~250MB)
CHUNKSIZE = 1_000_000

# Dictionary mapping work_id → [year, set of countries]
work_map: dict[str, tuple[int, set[str]]] = {}

# Read file in chunks and accumulate country sets per work
for chunk in pd.read_csv(
    FILENAME,
    usecols=COLS,
    chunksize=CHUNKSIZE,
    low_memory=True
):
    # iterate rows in chunk
    for wid, year, country in zip(chunk.work_id, chunk.publication_year, chunk.country):
        if wid in work_map:
            work_map[wid][1].add(country)
        else:
            # store publication_year and a set initialized with this country
            work_map[wid] = [year, {country}]

```

```python
# Prepare output rows: only works with >1 distinct country
rows: list[tuple[str, int, str]] = []
for wid, (year, countries) in work_map.items():
    if len(countries) > 1:
        # join countries with semicolon (or comma)
        country_list = ";".join(sorted(countries))
        rows.append((wid, year, country_list))

# Build DataFrame, sort by year then work_id, and write to CSV
out_df = pd.DataFrame(rows, columns=["work_id", "publication_year", "countries"])
out_df.sort_values(["publication_year", "work_id"], inplace=True)
out_df.to_csv(OUTPUT_FILE, index=False)

print(f"Saved {len(out_df)} works with multi-country authors to '{OUTPUT_FILE}'")
```
